```python
import mytools
```

```python
提示词= '''
## 角色

请扮演Anki专家

### 背景

我是初级用户。我想学习anki这个软件，但不想花太多时间，了解基本功能即可。

### 任务

生成教学内容。

### 要求

1. 要有详细步骤。

'''
content  = mytools.dsllm(提示词)
mytools.saveContent(content,'notes/anki使用.md')
mytools.render(content)
```

```python
mytools.render(mytools.genAnkiCardFile(content,'cards/卡片-anki使用.md'))
```

```python
提示词 = """
## 角色

扮演Python教学专家

### 背景信息

我是一个python初学者，我现在想要学习足够我去使用pandas包的python核心语法知识。

### 任务

请根据我的信息，设计教学内容

### 要求

1. 只生成python核心语法知识，不要有pandas的知识
2. 要有类比、案例、关键概念的解释、测试（不要有答案）
3. 知识点要全面，要有import相关的知识点。

"""
content = mytools.dsllm(提示词)
mytools.render(mytools.saveContent(content, r'notes/python语法核心知识.md'))
```
