```python
# 创建DataFrame
print("1. 创建DataFrame示例:")
data = {
    '姓名': ['张三', '李四', '王五', '赵六'],
    '年龄': [25, 30, 35, 28],
    '城市': ['北京', '上海', '广州', '深圳'],
    '工资': [10000, 15000, 12000, 20000]
}
df = pd.DataFrame(data)
print(df)
print("\n")

# 获取数据
print("2. 获取数据示例:")
# 获取单列
print("获取'年龄'列:")
print(df['年龄'])
print("\n获取多列:")
print(df[['姓名', '工资']])


# 增加数据
print("3. 增加数据示例:")
# 增加新列
df['奖金'] = [1000, 2000, 1500, 3000]
print("增加新列后的DataFrame:")
print(df)
# 增加新行
new_row = pd.Series(['陈七', 32, '武汉', 18000, 2500], index=df.columns)
df.loc[len(df)] = new_row
print("\n增加新行后的DataFrame:")
print(df)
print("\n")

# 修改数据
print("4. 修改数据示例:")
# 修改单个值
df.loc[0, '工资'] = 11000
# 修改整列
df['奖金'] = df['奖金'].apply(lambda x: x * 1.1)
print("修改后的DataFrame:")
print(df)
print("\n")

# 删除数据
print("5. 删除数据示例:")
# 删除列
df_temp = df.drop('奖金', axis=1)
print("删除'奖金'列后的DataFrame:")
print(df_temp)
# 删除行
df_temp = df.drop(0, axis=0)
print("\n删除第1行后的DataFrame:")
print(df_temp)
```

```python
import pdfplumber

# 打开PDF文件
with pdfplumber.open('data/2022计算机设计大赛国奖.pdf') as pdf:
    # 获取第四页
    page = pdf.pages[5]
    
    # 提取表格
    tables = page.extract_tables()
    
    # 如果有表格，转换第一个表格为DataFrame
    if tables:
        df4 = pd.DataFrame(tables[0][1:], columns=tables[0][0])
        print("提取的表格内容:")
        print(df4)
```

```python
# 将DataFrame保存为Excel文件
df4.to_excel('data/获奖.xlsx', index=False)
```
